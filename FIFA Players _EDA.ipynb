```python
df.corr(numeric_only=True)['Overall'].sort_values(ascending=False)
```

```python
#visualising the correlation btw overall rating and players attritbute
import matplotlib.pyplot as plt
plt.figure(figsize=(10,15))

corr_overall = df.corr(numeric_only=True)['Overall'].sort_values()

corr_overall.plot(kind='barh', figsize=(10,15))

plt.title("Correlation with Overall Rating")
plt.xlabel("Correlation")
plt.ylabel("Attributes")
plt.show()
```

```python
#Visualising using seaborn to show the correlation btw overal and attributes of players 

import seaborn as sns
subset = df[['Overall','Reactions','Composure','Potential','Value ','Wage','ShortPassing','Vision']]
plt.figure(figsize=(10,8))
sns.heatmap(subset.corr(numeric_only=True), 
            annot=True, 
            cmap='coolwarm', 
            linewidths=0.5)
plt.title("Correlation Heatmap for Key Attributes")
plt.show()

```
